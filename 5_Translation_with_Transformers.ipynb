RNN block which represents words-- every block needs to be processed before passing info to next block. Here Serialization is happening which makes to slow to train. This is avoided in Transformer which does processing in Parallel + less time to train. Keep checking **diagram** .

1. In English Block 5 blocks. French 5 blocks. 2nd block in french <--> 1st block in English. 1st block will get highest weight of eg 0.95 rest 0.1. Encoder has key and value while query from Decoder. Until now exactly same as attention
2. Only we use intra Attention on Encoder ie Key, Value from Encoder and Query from encoder itself is passed. Here we are finding how similar prev word is to the next word in Encoder. Eg [do][u][play][soccer][pad]. How similar do is with all words. How similar u is with all other words and so on..... We get a confusion matrix type matrix with weights. Diagonals will have obviously higher weights. Here idea is play and soccer may have high correlation but in any other sentence block play and soccer may have very low correlation. Earlier we use to use Word Embedding(word2vec). Here we understand sentence as a whole.

3. In Encoding Assume (batch,sequence_length,embedding dims)--> (batch,5,32)--> (5,32).  (5,32) will be Query,Key,Value all having same dims. Now we do Query*(Key**T) ie (5,32)*(32,5)--> (5,5) which is the matrix with weights mentioned above

4. QK^T is now (5,5) matrix. root_of(dk) is embedding dims ie 32. We apply softmax --> softmax(QK^T/root_of(dk))  to get values in above (5,5) matrix.

5. After we get (5,5) , we multiply by Value(intra) --> shape(5,32) --> (5,5)*(5,32)--> (5,32). This is called Scaled Dot-Product Attention. Here we got rid of RNN and enables us Parallel Computation. Now we have (5,32) ie Value(inter)
(5,32)--> 5 is Sequence_length, 32 is Embedding dims. Input of query,key of shape (5,32) and output of shape (5,32) for 1 Scaled Dot-Product Attention.

6. Many such Scaled Dot-Product Attention togehter is called Multi-Head Attention. It allows the model to jointly attend to information from different representation subspaces at different positions. With single attenstion head ie single Scaled Dot-Product Attention, averaging inhibits this

7. If Multi-Head Attention is 4. Check diagram Capture.jpg

8. All outputs of Scaled Dot-Product Attention will be concat. So 1 output of (5,32) for 4 Scaled Dot-Product Attention will be (5,128). This is again passed in linear layer to again get (5,32) dims.

9. In Transformer Encoder, We have Add&Norm. Input Embedding will output shape of (5,32) which will be added to Step 8 output of shape (5,32) and then Normalized using Layer Norm instead of Batch Norm

10. Output of Step 9 is passed into Feed Forward(Forward Prop). Shape of (5,32) for 4 Scaled Dot-Product Attention will be (5,128). Again passed into Linear Layer to get shape(5,32)

11. Output of Step 9 and Step 10 will be added to get a shape of (5,32)

12. For a single Encoder block, we have 8 multi head attention heads. Nx tells how much time single Encoder block is repeated in Series ie One above the other. If Nx=6, we have 8*6=48  multi head attention heads. Check Diagram image.png

13. After Input Embedding--> we have Positional Encoding. We have no Recurrence and no Convolution. In order our model knows order of sequence,we must inject some info about position of token of sequence.

14. From Input_Embedding we get (batch,Sequence_len,Embedding_dims)(batch,5,32). We add Positional Encoding of same dims of (batch,5,32) to Input_Embedding output. We get Positional Encoding using PE formula.

15. In PE Formula, pos is position of Embedding like 0,1,2... i are rows from bottom. i=0,i=1....i=dmodel-1. dmodel is dims of vector(Embedding_dims).

16. For even positions, we will use sin and for odd use cos for rows in Embedding_dims.

17. In Decoder, everything is Similar except we have **Masked** Multi-Head Attention. In Training, Encoder recieves English while Decoder receives French. During Testing, Decoder will not receive French words(outputs-check diagram) so  Decoder will relay on Embeddings from Encoder.

18. Shifted Right is prev word pattern + current word dims to predicted the output.

19. **Masked** Multi-Head Attention is same (5,5) dims like in Encoder except that target word will only depend on previous word + current word while remaining words are masked or made to 0. Goal here is how related a particular word token is other token. Check diagram:Capture2.JPG

20. K, V(inter) passed into Decoder Multi-Head Attention and Q is received from Decoder which is masked from prev steps. K,V(inter) represents input same like Attention and Q represents French language with [start] and [end] token.

21. Nx number will create Encoder and Decoder structures in parallel and outputs will be combined. Input shape and Output shape will be maintained. Then depending on vocab size, shape is now (batch,Sequence_len,vocab size)

# Modeling

## Embedding

Positional Encoding

```python
def positional_encoding(model_size,SEQUENCE_LENGTH):
  """
  PE formula will be used. Shape will be (Sequence_len,model_size ie embedding_dims)
  """
  output=[]
  for pos in range(SEQUENCE_LENGTH): # 64 different vectors with dims embedding_dims--> store it in output
    PE=np.zeros((model_size)) # defining 1st positional encoding matrix of zeroes of dim model_size or embedding_dims(32)--> [0,0,0....0]--> we will individual update every index
    for i in range(model_size): # to add values in every positional encoding matrix 32 times
      if i%2==0: # when position is even, using sin as mentioned in the paper
        PE[i]=np.sin(pos/(10000**(i/model_size)))
      else:
        PE[i]=np.cos(pos/(10000**((i-1)/model_size))) # to adjust 2i+1 in the paper, we use i-1
        # PE of [0,0,0...0] will be updated and appended to output[] at index[0] then 1,2,3,4,5 in our example
    output.append(tf.expand_dims(PE,axis=0))
  out=tf.concat(output,axis=0)
  out=tf.expand_dims(out,axis=0)
  return tf.cast(out,dtype=tf.float32)
```

###Input Embedding in Encoder and Output Embedding in Decoder--> To get Embedding Vector of length vocab size.

Positional Encoding from above code will be added which will have same dims as Embedding and this will be passed into Multi-layer Attention

```python
class Embeddings(Layer):
  def __init__(self, sequence_length, vocab_size, embed_dim,): # every word--> embedding vector of some dims + positional_encoding
    super(Embeddings, self).__init__()
    self.token_embeddings=Embedding( #(5,32) in our eg
        input_dim=vocab_size, output_dim=embed_dim) # input and output
    self.sequence_length = sequence_length
    self.vocab_size = vocab_size
    self.embed_dim = embed_dim

  def call(self, inputs):
    embedded_tokens = self.token_embeddings(inputs) # words in English/French will be inputs --> embedded_tokens are vectors
    embedded_positions=positional_encoding(
        self.embed_dim,self.sequence_length) # putting values for positional embedding from above code
    return embedded_tokens + embedded_positions # combining our Embedding + positional_embedding. This step represents + in Encoder/Decoder

  def compute_mask(self, inputs, mask=None): # how much every token is related to one another
    return tf.math.not_equal(inputs, 0)
    # here non important words like pad in row and pad in col- weights are zero, mask eg above
    # here eg (5,5) will be holding important values while pad in rows and cols will be 0
    #This is passed in Multi-head Attention
```

## Custom MultiHeadAttention

jump to TransformerEncoder --------------->




```python
class CustomSelfAttention(Layer): # Scaled Dot-Product Attention for personal self attention, check diagram
  def __init__(self,model_size):
    super(CustomSelfAttention,self).__init__()
    self.model_size=model_size
  def call(self,query,key,value,masking):
    ######## compute scores
    score=tf.matmul(query,key,transpose_b=True) #QK**T
    ######## scaling
    score/=tf.math.sqrt(tf.cast(self.model_size,tf.float32)) #QK**T/root(d_model eg 32)
    print(score)

    ######## masking
    masking=tf.cast(masking,dtype=tf.float32) #pad masking, check diagram
    print(masking)
    score+=(1.-masking)*-1e10 #making pad value very high -ve value so that softmax will consider them 0 or close to 0. We reverse 0<-->1 so that -ve values can be applied to padding and it becomes a large -ve value
    print(score) #padding applied

    ######## attention_weights
    #Run the check as pad values are same
    print(tf.nn.softmax(score,axis=-1))
    #to make pad values 0 we multiple by masking
    attention=tf.nn.softmax(score,axis=-1)*masking # attention weights will only be shared in values with 0 as pad values are high -ve value
    #softmax will make pad values 0
    print(attention)
    ######## output
    head=tf.matmul(attention,value)
    return head

#Assume we get (64,64)
```

```python
class CustomMultiHeadAttention(Layer): #Multi Head Attention
  def __init__(self,num_heads,key_dim): #key_dim is d_model eg 32
    super(CustomMultiHeadAttention,self).__init__()
    """
    head1 = Attention(QWi,KWi,VWi) every head will have a Linear Dense layer
    MultiHead(Q,K,V) = [head1,head2..headn]Wi. If MultiHead is 8 we have 8 Linear Dense Layer
    """
    self.num_heads=num_heads
    self.dense_q=[Dense(key_dim) for _ in range(num_heads)]
    self.dense_k=[Dense(key_dim) for _ in range(num_heads)]
    self.dense_v=[Dense(key_dim) for _ in range(num_heads)]
    self.dense_o=Dense(key_dim) #Multi-Head Attention top Linear Layer -->

    self.self_attention=CustomSelfAttention(key_dim)

  def call(self,query,key,value,attention_mask): #attention_mask from prev step
    heads=[] # Scaled Dot-Product Attention h (check the diagram)

    for i in range(self.num_heads): #Eg-8, dense linear layer of q[0],k[0],v[0], mask --> attention-->head[0]-->append-->repeat
        head=self.self_attention(self.dense_q[i](query),
                                 self.dense_k[i](key),
                                 self.dense_v[i](value),
                                 attention_mask)
        heads.append(head)
    heads=tf.concat(heads,axis=2) #concat layer in diagram

    heads=self.dense_o(heads) #Linear Layer
    return heads

#We can use this instead of tensorflow Multi-Head Attention
# Now our dims will be eg (8,64,64)
```

```python
class TransformerEncoder(Layer):
    def __init__(self, embed_dim, dense_dim, num_heads,):# mask made in embedding layer will be auto passed which will be boolean
        super(TransformerEncoder, self).__init__()
        self.embed_dim = embed_dim
        self.dense_dim = dense_dim
        self.num_heads = num_heads
        self.attention = MultiHeadAttention( #CustomMultiHeadAttention
            num_heads=num_heads, key_dim=embed_dim, #key_dim is d_model or embedding_vector eg-32, num_heads=5 in eg
                                            )

        #Feed Forward layer defined here
        self.dense_proj=tf.keras.Sequential(
                                            [Dense(dense_dim, activation="relu"),
                                             Dense(embed_dim),]
                                            ) #shape of(5,32) will change and again we get (5,32)
        #Add & Norm Norm layer - 1
        self.layernorm_1 = LayerNormalization()
        #Add & Norm Norm layer - 2 ie in diagram the top layer of Encoder
        self.layernorm_2 = LayerNormalization()

        self.supports_masking = True

    def call(self, inputs, mask=None):
      if mask is not None:
        mask1 = mask[:, :, tf.newaxis]
        mask2 = mask[:,tf.newaxis, :]
        padding_mask = tf.cast(mask1&mask2, dtype="int32") #boolean mask recieved received changed to int32

      attention_output = self.attention(
          query=inputs, key=inputs,value=inputs,attention_mask=padding_mask
                                        )
      #From diagram-> Embedding which is added to Add&Norm
      proj_input = self.layernorm_1(inputs + attention_output)
      proj_output = self.dense_proj(proj_input)
      return self.layernorm_2(proj_input + proj_output)
```

## Decoder

```python
class TransformerDecoder(Layer):
  def __init__(self, embed_dim, latent_dim, num_heads,):
    super(TransformerDecoder, self).__init__()
    self.embed_dim = embed_dim
    self.latent_dim = latent_dim
    self.num_heads = num_heads
    self.attention_1=MultiHeadAttention( #CustomMultiHeadAttention
        num_heads=num_heads, key_dim=embed_dim
    )
    self.attention_2=MultiHeadAttention(
        num_heads=num_heads, key_dim=embed_dim
    )
    self.dense_proj = tf.keras.Sequential(
        [Dense(latent_dim, activation="relu"),Dense(embed_dim),]
    )
    self.layernorm_1=LayerNormalization()
    self.layernorm_2=LayerNormalization()
    self.layernorm_3=LayerNormalization()
    self.supports_masking = True


  def call(self, inputs, encoder_outputs, mask=None):

    if mask is not None: #We have a masked Attention layer
      mask1 = mask[:, :, tf.newaxis]
      mask2 = mask[:,tf.newaxis, :]

      padding_mask = tf.cast(mask1&mask2, dtype="int32")

      causal_mask=tf.linalg.band_part(tf.ones([tf.shape(inputs)[0],
                                               tf.shape(inputs)[1], #lower traingle to be 1's, shape of traingle from inputs
                                               tf.shape(inputs)[1]
                                               ],dtype=tf.int32),-1,0) #if shape is (2,5,32)--we want (2,5,5). So 2 is [0], 5 is [1] and 5 is [1]

      # padding_mask is where pad values are 0, causal mask is lower traingle
      combined_mask=tf.minimum(padding_mask,causal_mask) #this will be passed further in normal Multi-head Attention

    # For Masked(causal) Multi-Head Attention, causal mask diagram  below
    attention_output_1 = self.attention_1(
        query=inputs,key=inputs,value=inputs,
        attention_mask=causal_mask

                                        )
    out_1 = self.layernorm_1(inputs + attention_output_1) #1st Add&Norm --> out_1 has 2 splits- 1 in Multi-head Attention as Query, other in next Add&Norm

    #Multi-head_attention, key,value received from Encoder, Query from out_1
    attention_output_2,scores= self.attention_2(
        query=out_1,key=encoder_outputs,value=encoder_outputs,
        attention_mask=combined_mask,
        return_attention_scores=True #attention weights will be outputed
                                              )
    out_2 = self.layernorm_2(out_1 + attention_output_2) #2nd Add&Norm output

    proj_output = self.dense_proj(out_2) #Feed Forward aka Dense
    return self.layernorm_3(out_2 + proj_output),scores #3rd Add&Norm
```

## Transformer Model

Bring it all together

```python
EMBEDDING_DIM=128 #D_model eg 32
D_FF=1024 #Dense layers
NUM_HEADS=8
NUM_LAYERS=4 #Nx from diagram.
NUM_EPOCHS=20
attention_scores={}
```

```python
encoder_inputs=Input(shape=(None,), dtype="int64", name="input_1")
x = Embeddings(ENGLISH_SEQUENCE_LENGTH,VOCAB_SIZE,EMBEDDING_DIM)(encoder_inputs)

for _ in range(NUM_LAYERS):
  x=TransformerEncoder(EMBEDDING_DIM,D_FF,NUM_HEADS)(x)
encoder_outputs=x #Key, Value output from Encoder


decoder_inputs=Input(shape=(None,), dtype="int64", name="input_2")

x = Embeddings(FRENCH_SEQUENCE_LENGTH,VOCAB_SIZE,EMBEDDING_DIM)(decoder_inputs)

for i in range(NUM_LAYERS):
  x,scores=TransformerDecoder(EMBEDDING_DIM,D_FF,NUM_HEADS)(x, encoder_outputs) #enccoder output in Decoder multihead Attention, x is Add#Norm query

  attention_scores[f'decoder_layer{i+1}_block2']=scores # For Decoder Multi-Head Attention

# We want to change is shape from (batch,sequence_length,d_model) --> (batch,sequence_length,vocab_size), we have linear layer, passed into softmax
decoder_outputs=Dense(VOCAB_SIZE, activation="softmax")(x)


attention_score_model= tf.keras.Model(
    [encoder_inputs, decoder_inputs], attention_scores, name="atttention_score_model"
)
transformer = tf.keras.Model(
    [encoder_inputs, decoder_inputs], decoder_outputs, name="transformer"
)
transformer.summary()
```

# Training

```python
class BLEU(tf.keras.metrics.Metric):
    def __init__(self,name='bleu_score'):
        super(BLEU,self).__init__()
        self.bleu_score=0

    def update_state(self,y_true,y_pred,sample_weight=None):
      y_pred=tf.argmax(y_pred,-1)
      self.bleu_score=0
      for i,j in zip(y_pred,y_true):
        tf.autograph.experimental.set_loop_options()

        total_words=tf.math.count_nonzero(i)
        total_matches=0
        for word in i:
          if word==0:
            break
          for q in range(len(j)):
            if j[q]==0:
              break
            if word==j[q]:
              total_matches+=1
              j=tf.boolean_mask(j,[False if y==q else True for y in range(len(j))])
              break

        self.bleu_score+=total_matches/total_words

    def result(self):
        return self.bleu_score/BATCH_SIZE
```

```python
transformer.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(2e-4),)
    #metrics=[BLEU()],
    #run_eagerly=True)
```

```python
def translator(english_sentence):
  tokenized_english_sentence=english_vectorize_layer([english_sentence])
  shifted_target='starttoken'

  for i in range(FRENCH_SEQUENCE_LENGTH):
    tokenized_shifted_target=french_vectorize_layer([shifted_target])
    output=transformer.predict([tokenized_english_sentence,tokenized_shifted_target])
    french_word_index=tf.argmax(output,axis=-1)[0][i].numpy()
    current_word=index_to_word[french_word_index]
    if current_word=='endtoken':
      break
    shifted_target+=' '+current_word
  return shifted_target[11:]
```

```python
class Transformer(tf.keras.Model):
  def __init__(self,transformer):
    super(Transformer,self).__init__()
    self.transformer=transformer

  def compile(self,loss_fn,optimizer):
    super(Transformer,self).compile()
    self.optimizer=optimizer
    self.loss_fn=loss_fn
    self.loss_metric=tf.keras.metrics.Mean(name='loss')

  @property
  def metrics(self):
    return [self.loss_metric,]

  def train_step(self,x_y):
    inputs,target=x_y
    encoder_input=inputs['input_1']
    shifted_target=inputs['input_2']

    with tf.GradientTape() as recorder:

      output,_=self.transformer([encoder_input,shifted_target])
      loss=self.loss_fn(target,output)

    partial_derivatives = recorder.gradient(loss,self.transformer.trainable_weights)
    self.optimizer.apply_gradients(zip(partial_derivatives, self.transformer.trainable_weights))

    self.loss_metric.update_state(loss)

    return {'loss':self.loss_metric.result()}
```
